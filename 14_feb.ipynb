Multithreading in Python
What is Multithreading?
Multithreading in Python refers to the concurrent execution of multiple threads (smaller units of a process) within a single process. It allows for multiple operations to be performed simultaneously within a single application.

Why is it Used?
Multithreading is used to improve the performance of applications by allowing multiple threads to run in parallel. It is particularly useful for I/O-bound tasks, such as reading from a file or making network requests, as these tasks can be performed while other operations continue.

Module Used
The module used to handle threads in Python is called the threading module.

Why Use the Threading Module?
The threading module is used because it provides a high-level interface for working with threads. It allows for easy creation, control, and management of threads in a Python program.

Python Program to Create Two Threads





# t
import threading

def print_squares():
    for i in range(1, 6):
        print(f"Square of {i}: {i ** 2}")

def print_cubes():
    for i in range(1, 6):
        print(f"Cube of {i}: {i ** 3}")










# Create threads
thread1 = threading.Thread(target=print_squares)
thread2 = threading.Thread(target=print_cubes)

# Start threads
thread1.start()
thread2.start()

# Wait for threads to complete
thread1.join()
thread2.join()
Advantages and Disadvantages of Multithreading
Advantages:

Improved performance and concurrency.
Efficient utilization of CPU resources.
Simplified design for some applications (e.g., GUI applications).
Disadvantages:

Complexity in code, leading to potential difficulties in debugging.
Potential for race conditions and deadlocks.
Python's Global Interpreter Lock (GIL) can limit the effectiveness of CPU-bound multithreading.
Deadlocks and Race Conditions
Deadlocks:
A deadlock occurs when two or more threads are blocked forever, waiting for each other. This usually happens when two threads have a circular dependency on a pair of resources.

Race Conditions:
A race condition occurs when two or more threads can access shared data and try to change it at the same time. The outcome depends on the non-deterministic ordering of thread execution, leading to unpredictable results.

Example of Deadlock
python
Copy code
import threading

lock1 = threading.Lock()
lock2 = threading.Lock()

def thread1():
    with lock1:
        with lock2:
            print("Thread 1 has acquired lock1 and lock2")

def thread2():
    with lock2:
        with lock1:
            print("Thread 2 has acquired lock2 and lock1")

t1 = threading.Thread(target=thread1)
t2 = threading.Thread(target=thread2)

t1.start()
t2.start()

t1.join()
t2.join()
In this example, thread1 acquires lock1 and waits for lock2, while thread2 acquires lock2 and waits for lock1, causing a deadlock.

Example of Race Condition
python
Copy code
import threading

counter = 0
lock = threading.Lock()

def increment():
    global counter
    for _ in range(100000):
        with lock:
            counter += 1

threads = []
for _ in range(10):
    t = threading.Thread(target=increment)
    t.start()
    threads.append(t)

for t in threads:
    t.join()

print(counter)
Without the lock, multiple threads could increment the counter simultaneously, leading to incorrect results due to race conditions.
